# Procesamiento

Si vemos las gr치ficas anteriores nos muestran la acumulaci칩n de casos de COVID en el tiempo, sin embargo no nos permiten describir el comportamiento d칤a a d칤a de lo que est치 sucediendo con los contagios y las muertes. 
Por lo tanto, realizaremos un procesamiento en el cual calcularemos la diferencia entre los valores diarios para tener informaci칩n relevante de la tendencia diaria de los contagios y las muertes. 

Adicionalmente, como son muchos valores en el tiempo podemos realizar un filtro d media m칩vil, lo cual nos va a permitir suavizar la se침al y tener una primera aproximaci칩n sobre como ser치 el comportamiento de la se침al en el tiempo. Es decir, asumiremos que el valor futuro de nuestras variables depende de sus $k$ valores anteriores.

````python
def daily_increase(data):
    d = [] 
    for i in range(len(data)):
        if i == 0:
            d.append(data[0])
        else:
            d.append(data[i]-data[i-1])
    return d 

def maFilter(x, N):   # N ventana de tiempo                        
  xTemp = np.zeros([len(x) + (N-1)])
  xTemp[(N-1):] = x
  y = np.zeros(len(x))
  
  for i in range(0, len(x)):
    y[i] = xTemp[i:i +N]. sum() / float(N)
  return y
````

````python
#Valor de la ventana. Este 7 significa 7 d칤as. Vamos a hacerle un filtro por semana
window = 7
#Cases
world_daily_increase = daily_increase(world_cases_df.cases)
world_daily_increase_avg = maFilter(world_daily_increase, window)

# deaths
world_daily_death = daily_increase(total_deaths_df.deaths)
world_daily_death_avg = maFilter(world_daily_death, window)
````

## Calidad del pron칩stico

Para realizar un adecuado pron칩stico es importante tener en cuenta las m칠tricas que nos permitiran estimar si el pron칩stico que se realiz칩 es adecuado

* **$R^2$**: coeficiente de determinaci칩n (en econometr칤a, esto puede interpretarse como el porcentaje de varianza explicado por el modelo). Se puede calcular utilizando la funci칩n `sklearn.metrics.r2_score`

* **Error absoluto medio ($MAE$)**: se trata de una m칠trica interpretable de la medida de la diferencia entre dos variables continuas. `sklearn.metrics.mean_absolute_error`

* **Mediana del error absoluto ($MedAE$)**: es una medida robusta para los valores at칤picos. Se calcula tomando la mediana de todas las diferencias absolutas entre el objetivo y la predicci칩n.  
`sklearn.metrics.median_absolute_error`

* **Error cuadr치tico medio ($MSE$)**: es una m칠trica de riesgo correspondiente al valor esperado del error o la p칠rdida cuadr치tica `sklearn.metrics.mean_squared_error`

* **Error logar칤tmico medio al cuadrado ($MSLE$)**: es lo mismo que MSE, pero tomamos el logaritmo de la serie. Como resultado, damos m치s peso a los peque침os errores tambi칠n. Esto se utiliza generalmente cuando los datos tienen tendencias exponenciales. `sklearn.metrics.mean_squared_log_error`

* **Error de porcentaje absoluto medio ($MAPE$)**: esto es lo mismo que $MAE$ pero se calcula como porcentaje, lo que es muy conveniente cuando se quiere explicar la calidad del modelo.

```
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100```


````python
from sklearn.metrics import r2_score, median_absolute_error, mean_absolute_error
from sklearn.metrics import median_absolute_error, mean_squared_error, mean_squared_log_error

def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
````

````python
#Funci칩n para graficar la media m칩vil
#La funci칩n rolling nos ayuda a c치lcular la media m칩vil
def plotMovingAverage(series, window, plot_intervals=False, scale=1.96, plot_anomalies=False):
    rolling_mean = series.rolling(window=window).mean()
    plt.figure(figsize=(10,5))
    plt.title("Media m칩vil\n Tama침o de la ventana = {}".format(window))
    plt.plot(rolling_mean, "g", label="Tendencia media m칩vil")

    # Graficar intervalos de confianza para valores suavizados
    if plot_intervals:
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        deviation = np.std(series[window:] - rolling_mean[window:])
        lower_bond = rolling_mean - (mae + scale * deviation)
        upper_bond = rolling_mean + (mae + scale * deviation)
        plt.plot(upper_bond, "r--", label="Intervalos superior / Inferior")
        plt.plot(lower_bond, "r--")
        
        # Encontrar valores anormales
        if plot_anomalies:
            anomalies = pd.DataFrame(index=series.index, columns=series.columns)
            anomalies[series<lower_bond] = series[series<lower_bond]
            anomalies[series>upper_bond] = series[series>upper_bond]
            plt.plot(anomalies, "ro", markersize=10)
        
    plt.plot(series[window:], label="Valor actual")
    plt.legend(loc="upper left")
    plt.grid(True)
````

## Otros modelos
En este caso se puede observar que el suavizado de media m칩vil permite capturar la estacionalidad mensual.
Sin embargo, podemos realizar otras pruebas considerando modelos m치s complejos.

Para esto se puede calcular el utilizando el Promedio ponderado que es una modificaci칩n de la media m칩vil, donde se establece un peso de ponderaci칩n para realizar el c치lculo promedio.

Tambi칠n un m칠todo interesante es el suavizado exponencial, donde el valor del modelo es un promedio ponderado entre el valor verdadero actual y los valores del modelo anterior. El peso  $풤$  se llama factor de suavizaci칩n. Define cu치n r치pidamente "olvidaremos" la 칰ltima observaci칩n verdadera disponible. Cuanto m치s peque침a es  $풤$ , m치s influencia tienen las observaciones anteriores y m치s suave es la serie.

La exponencialidad se oculta en la recursividad de la funci칩n -- multiplicamos por (1 - 풤 ) cada vez, que ya contiene una multiplicaci칩n por (1- 풤 ) de los valores del modelo anterior.

````python
#M칠todo para suavisar una serie usuando suavizado exponencial
def exponential_smoothing(series, alpha):
    """
        Par치metro de suavizado alpha - float [0.0, 1.0]
        series: Seria de datos que vamos a suavizar
    """
    result = [series[0]] # El primer valor es el mismo de la serie
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    return result
````

````python
def plotExponentialSmoothing(series, alphas):
    """
    Gr치fica el suavizado exponencial con diferentes alphas        
    """
    plt.figure(figsize=(10, 5))
    plt.plot(series.values, "c", label = "Actual")
    for alpha in alphas:
        plt.plot(exponential_smoothing(series, alpha), label="Alpha {}".format(alpha))
    plt.legend(loc="best")
    plt.axis('tight')
    plt.title("Suavizado exponencial")
    plt.grid(True);
````

## 游늳 Forecast

La idea del siguiente algoritmo es poder predecir m치s valores futuros de la serie temporal que estamos analizando. En este caso vamos a realizar un modelado de la serie a partir de su comportamiento pasado. 

Para esto vamos a utilizar la librer칤a de ```Scikit-learn``` espec칤ficamente vamos a utlizar el ```skforest``` que contienen las clases y funciones necesarias para adaptar cualquier modelo a problemas de forecasting

### Entrenamiento 

Para probar otros modelos, es importante definir como se realizar치 la estimaci칩n de los par치metros. 

Recordemos que siempre debemos elegir una funci칩n de p칠rdida adecuada para la tarea, la cual nos indicar치  lo cerca que se aproxima el modelo a los datos. 

Despu칠s utilizando la validaci칩n cruzada se evaluar치 la funci칩n de p칠rdida elegida para los par치metros del modelo, se calcular치 el gradiente y se ajustr치n nuevamente los par치metros.

El reto est치 en c칩mo hacer la validaci칩n cruzada, puesto que las series de tiempo tienen una estructura temporal y no se pueden mezclar los valores de una manera aleatoria. Esta es la raz칩n por lo que se debe utilizar un enfoque de optimizaci칩n del modelo. 

Para esto entonces se utilizar치 la **"validaci칩n cruzada sobre base continua"**

La idea es entrenar el modelo en un peque침o segmento de la serie temporal hasta un valor de $t$ y luego hacer predicciones para los pasos $t+ n$ y calcular el error. Despu칠s se ampliar치 el entrenamiento y se realizar치n predicciones de $t+n$ hasta $t+2n$ y se continuar치 moviendo el segmento de la prueba hasta alcanzar la 칰ltima observaci칩n disponible.

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

from skforecast.ForecasterAutoreg import ForecasterAutoreg
#Ensablar el bosque
from skforecast.model_selection import grid_search_forecaster
#Encontrar los parametros 칩ptimos del modelo

from joblib import dump, load # Para almacenar los modelos
import warnings

warnings.filterwarnings('ignore')
````

````python
#El indice del dataset debe ser las fechas o la variable temporal de la serie. Convertimos las fechas y los convertimos en el indice del dataset
world_daily_increase_df['dates'] = pd.to_datetime(world_daily_increase_df['dates'], format = '%m/%d/%y')
world_daily_increase_df = world_daily_increase_df.set_index('dates')
world_daily_increase_df.head()
````

````python
##Ajuste de hyperparametros. Esta funci칩n nos devolvera una matrix con el mejor p치rametros
def modelParameters(data, steps): 
    #Definimos el data de entrenamiento. Le quitamos los 칰ltimos 30 d칤as
    data_train = data[:-steps]
    #Definimos nuestro modelo
    forecaster = ForecasterAutoreg(
                    regressor = RandomForestRegressor(random_state=123),
                    lags      = 7)

    lags_grid = [5, 10, 15, 20]                           # ventana temporal

    param_grid = {'n_estimators': [50, 100, 500],         # n칰mero de 치rboles 
                  'max_depth': [3, 5, 10, 15]}

    #Con este c칩digo probamos todas la combinaciones posibles para obtener el m칤nimo error
    resultados_grid = grid_search_forecaster(
                            forecaster         = forecaster,
                            y                  = data_train['data'],
                            param_grid         = param_grid,
                            lags_grid          = lags_grid,
                            steps              = steps,
                            refit              = True,
                            metric             = 'mean_squared_error',
                            initial_train_size = int(len(data_train)*0.5),
                            fixed_train_size   = False,
                            return_best        = True,
                            verbose            = False
                       )
    return resultados_grid
````

````python
#Funci칩n para predecir el m칩delo
def predictModel(data, steps, max_depth, n_estimators, lags): 
    #Se separan los datos de entrenamiento de los datos de test
    data_train = data[:-steps]
    data_test  = data[-steps:] 
    regressor = RandomForestRegressor(max_depth = max_depth, 
                                      n_estimators = n_estimators, random_state=123)
    
    forecaster = ForecasterAutoreg(
                regressor = regressor,
                lags      = lags
             )

    forecaster.fit(y=data_train['data'])
    predictions = forecaster.predict(steps = steps)
    
    dfPred = pd.DataFrame(predictions)
    #Fechas de nuestro datatest
    dfPred.index = pd.date_range(start=data_test.index.min(), end=data_test.index.max())

    dfPred['Real']  = data_test['data'].values
    dfPred['Diference'] = np.abs((dfPred['Real'] - dfPred['pred']))
    print(mean_absolute_percentage_error(dfPred['Real'], dfPred['pred'] ), "%")
    
    return dfPred, forecaster
````

````python
def predGraph(dataTrain, dataTest, dataPred, zoom = 0):
    fig, ax = plt.subplots(figsize = (15, 5))
    dataTrain['data'][zoom:].plot(ax = ax, label = 'train')
    dataTest['data'].plot(ax = ax, label = 'test')
    dataPred['pred'].plot(ax = ax, label = 'predicciones')
    ax.legend();
    return
````
